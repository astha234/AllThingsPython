```python
#!usr/bin/python
import pandas as pd
import numpy as np
import xlsxwriter as xl
import csv
```

```python
xl1 = pd.ExcelFile("/home/asthaab/megha/Combined_US_Logistics_Spend.xlsx")
DataFrame1=xl1.parse('Combined Freight-Invoices')
len_megha=len(DataFrame1)
```

```python
DataFrame1.head()
len_megha
```

```python
count=1
workbook = xl.Workbook('Combined_US_Logistics_Spend_mod.xlsx',{'nan_inf_to_errors': True})
worksheet = workbook.add_worksheet('Combined Freight-Invoices')
for i in range(0,len_megha):
    row_i=DataFrame1.irow(i).real
    name_i=str(row_i[26])
    if (' ' in name_i) and (name_i!=''):
        name_i_split=name_i.split(' ')
        name_i_split=[x for x in name_i_split if x != '']
        print i,name_i, name_i_split,len(name_i_split)
        for j in range(0,len(name_i_split)):
            for k in range(0,26):
                worksheet.write(count,k,str(row_i[k]))
            worksheet.write(count,26,name_i_split[j])
            count+=1
    else:
        for k in range(0,27):
            worksheet.write(count,k,str(row_i[k]))
        count+=1
            
workbook.close()
```
